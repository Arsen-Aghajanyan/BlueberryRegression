```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_train = df_train.drop(['id', 'Row#', 'MaxOfUpperTRange', 'MinOfUpperTRange', 'AverageOfUpperTRange', 'MaxOfLowerTRange', 'MinOfLowerTRange', 'AverageOfLowerTRange'],axis=1)
df_test = df_test.drop(['id', 'Row#', 'MaxOfUpperTRange', 'MinOfUpperTRange', 'AverageOfUpperTRange', 'MaxOfLowerTRange', 'MinOfLowerTRange', 'AverageOfLowerTRange'],axis=1)
```

```python
X_train = df_train.drop('yield', axis=1)
y_train = df_train['yield']
X_test = df_test.copy()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
model = RandomForestRegressor()
scaler = MinMaxScaler()
```

```python
pipe = Pipeline([('scaler', scaler), ('model', model)])
```

```python
params = {
    'model__n_estimators':[100],
    'model__criterion':["squared_error", "absolute_error"],
    'model__max_depth':[None, 4, 8],
    'model__max_features':[None, 4, 8]
    
}
```

```python
full_mod = GridSearchCV(pipe, param_grid=params, cv=3,verbose=2)
```

```python
full_mod.fit(X_train, y_train)
```

```python
y_pred = full_mod.predict(X_test)
```
